```python
df1 = pd.DataFrame({
    "ID" : [1,2,3,4],
    "Name" : ["Alice","Bob","Charlie","David"],
    "Subject" : ["Math","English","Science",'Arts']
})

df2 = pd.DataFrame({
    "ID" : [3,4,5,6],
    "Score" : [90,85,80,95],
    "Grade" : ["A","B","C","A"]
})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on='ID',how='inner')
```

```python
pd.merge(df1,df2,on='ID',how='outer')
```

```python
pd.merge(df1,df2,on='ID',how='left')
```

```python
pd.merge(df1,df2,on='ID',how='right')
```

```python
pd.merge(df2,df1,on='ID',how='left')
```

```python
pd.merge(df2,df1,on='ID',how='right')
```

```python
df3 = pd.DataFrame({
    "Student_ID" : [1,2,3,4],
    "Name" : ["Alice","Bob","Charlie","David"],
    "Subject" : ["Math","English","Science",'Arts']
})

df4 = pd.DataFrame({
    "ID" : [3,4,5,6],
    "Score" : [90,85,80,95],
    "Grade" : ["A","B","C","A"]
})
```

```python
df3
```

```python
df4
```

```python
pd.merge(df3,df4,left_on='Student_ID',right_on='ID',how='inner')
```
